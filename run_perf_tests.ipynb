# Run the emulators with various entries and record the results

```python
import os
import numpy as np
```

```python
algorithms = ['lru', 'fifo', 'newalg']
page_sizes = [2 ** x for x in range(1, 7)] # 2 to 64
memory_sizes = [2** x for x in range(7, 15)] # 128 to 16384
fixtures_dir = 'test/fixtures'
test_result_dir = 'results'
test_fixtures = ['compressor.log', 'matriz.log', 'compilador.log', 'simulador.log']
```

```python
def prepare_runs(test_result_dir, algorithms):
    os.system("make clean<<EOF\ny\nEOF")
    os.system("make unitTest")
    os.system("make")
    for algorithm in algorithms:
        os.system("mkdir -p {}/{}".format(test_result_dir, algorithm))

def run_and_record(algorithm, fixtures_dir, test_fixture, page_size, memory_size):
    cmd = "./build/main.out {} {} {} {} {} 1>{}".format(
        algorithm, fixtures_dir + '/' + test_fixture, page_size, memory_size,
        "test",
        f"{test_result_dir}/{algorithm}/{test_fixture}_{page_size}_{memory_size}",
    )
    print(cmd)
    os.system(cmd)
    
def prepare_and_run_all(test_result_dir, fixtures_dir, algorithms, test_fixtures, page_sizes, memory_sizes):
    prepare_runs(test_result_dir, algorithms)
    for algorithm in algorithms:
        for test_fixture in test_fixtures:
            for page_size in page_sizes:
                for memory_size in memory_sizes:
                    run_and_record(algorithm, fixtures_dir, test_fixture, page_size, memory_size)
```

```python
prepare_and_run_all(test_result_dir, fixtures_dir, algorithms, test_fixtures, page_sizes, memory_sizes)
```
